```
X=data.iloc[:,[0,1,2,3]].values
wcss=[]
for i in range(1,20):
  kmeans=KMeans(random_state=0,n_clusters=i)
  kmeans.fit(X)
  print(kmeans.inertia_)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,20),wcss)
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```
#so number of clusters will be taken as 3
kmean=KMeans(n_clusters=3,random_state=0)
y_pred=kmean.fit_predict(X)
print(y_pred)
```

```
centroids=kmean.cluster_centers_
print(centroids)
```

```
#now plotting of the clusters
plt.scatter(X[y_pred==0,0],X[y_pred==0,1],s=50,c='red',label='Iris-setosa')
plt.scatter(X[y_pred==1,0],X[y_pred==1,1],s=50,c='blue',label='Iris-setosa')
plt.scatter(X[y_pred==2,0],X[y_pred==2,1],s=50,c='green',label='Iris-versicolor')
plt.scatter(centroids[:,0],centroids[:,1],s=50,c='black',label='Centroid')
plt.xlabel('Sepal Length in cm')
plt.ylabel('Sepal Width in cm')
plt.title('Clusters for Sepal Length and Sepal Width')
plt.legend()
plt.show()
```

```
plt.scatter(X[y_pred==0,2],X[y_pred==0,3],s=50,c='red',label='Iris-setosa')
plt.scatter(X[y_pred==1,2],X[y_pred==1,3],s=50,c='blue',label='Iris-setosa')
plt.scatter(X[y_pred==2,2],X[y_pred==2,3],s=50,c='green',label='Iris-versicolor')
plt.scatter(centroids[:,2],centroids[:,3],s=50,c='black',label='Centroid')
plt.xlabel('Petal Length in cm')
plt.ylabel('Petal Width in cm')
plt.title('Clusters for Petal Length and Petal Width')
plt.legend()
plt.show()
```
